The first section is importing all of the necessary libraries for data visualization and machine learning. **matplotlib.pyplot** is a module mostly called by name plt. That is a very popular Python library for creating line plots, scatter plots and histograms. The module provides an interface, almost like **MATLAB**, for an easier use for data visualization. The second import statement is to load the **load_iris** function from **sklearn.datasets**, a part of the **scikit-learn** library, a powerful machine learning library. The load_iris function fetches the famous Iris dataset, comprising 150 samples of iris flowers from three species: **Setosa, Versicolor, and Virginica.** Every sample contains four numerical features: sepal length, sepal width, petal length, and petal width, all of which are useful for classification tasks. This dataset has mainly been used for **machine learning research** and tutorial due to its **well-structured information** and **well-defined boundaries** for the classification.


```python
# YORO, Leonardo Antero SP. - 302I
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

This section of the code loads the Iris dataset by using the **load_iris()** function from scikit-learn. The components of the dataset are stored in a variable called data, which is acting like a dictionary-like object, containing different components, such as **feature data, target labels, and feature names**, along with other metadata. Loading this dataset allows the program to use it later for **visualization, analysis, or machine-learning tasks**.

```python
# Load Data
data = load_iris()
```

The following piece of code pulls the feature matrix denoted as **X**, along with the target labels **y** from the iris dataset that has been loaded already. Here, **data.data** is assigned to the variable X, which contains the **numerical measurements** of iris flowers. This is a 2D array, in which every row represents a sample, and each column is related to one of the four features: sepal length, sepal width, petal length, and petal width. The variable y takes the value of  **data.target** that contains target values of labels for classification. Each value in y represents a category, where **0 represents Setosa**, **1 represents Versicolor**, and **2 represents Virginica**. The dataset is now set for **further analysis, visualization or machine learning training**, after assigning features and the target value as above.

```python
# Separate features and target
X = data.data
y = data.target
```

This section of the code splits the dataset with the **train_test_split** function from **sklearn.model_selection**. In machine learning models, the **data for training needs to be separated from evaluation** so that the generalized outcome produces good results in unseen data. The train_test_split function splits a dataset into two subsets, for **training X_train** and **y_train**, while the other is for **testing X_test and y_test**, which are divided with randomness. The **test_size=0.2** parameter indicates that 2**0% of the data will be used for testing** and **80% will be used to train the model**. The **random_state=0** makes sure that this **split is identical every time so that the run can be repeated**. This step is important during model development since **it prevents overfitting by checking the model's performance** on data it has never seen before during deployment.

```python
#Splitting the data for training and testing
from sklearn.model_selection import train_test_split
X_Train, X_Test, y_Train, y_Test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

This part initializes a **k-Nearest Neighbors classifier** using the **KNeighborsClassifier** class from **sklearn.neighbor**s. This algorithm is a very simple but incredibly powerful machine learning technique for classification and regression. The basic idea behind this type of algorithm is to find the **k closest data points to a given input** and assign the most common class of them as the predicted label. For the classifier here, **n_neighbors is set at 5** so it would predict for the **five nearest neighbors**. Generally, setting k=5 has been chosen to be the default as this has the **tendency to smoothen the noise** yet **preserves the decision** boundary. In training this classifier with the data, it would find the underlying pattern and, when new points were given to the classifier, use its knowledge in deciding on its category.

```python
#Create a KNeighbordsClassifier object
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
```

This fits the k-Nearest Neighbors model on the training data. So here, it applies its `fit` method of its `KNeighborsClassifier` class taking `X_Train` as the feature values of the training set and `y_Train` as the respective target labels. It doesn't do any explicit parameter learning at this step such as that done by many other machine learning algorithms, but `instead memorizes the data`. Instead, k-NN remembers the training samples so when a new data point is provided, `it could compute the distance to the nearest training points` and classify the new instance `based on the majority vote of its k closest neighbors`. Fitting the model to the training data makes it ready for predicting new or unseen data.

```python
#Train the model using the training data
knn.fit(X_Train, y_Train)
```

This section of the code applies the trained `k-Nearest Neighbors (k-NN) model` to make predictions on the test dataset. The predict method of the KNeighborsClassifier is called on `X_Test`, with the feature values of the test set contained inside it. The model, having already been trained on the training data, `now evaluates each test sample by comparing its features to those in the training set` and `determining the most common class among the nearest neighbors`. The resulting predicted class labels are `stored in the variable predictions`. This step is a necessary part of a performance evaluation, with the ability to `compare the prediction labels with actual labels y_Test coming from test set`.

```python
#Use the model to predict the test set
predictions = knn.predict(X_Test)
```

This section of code calculates the model's accuracy after it's trained to implement k-NN. Accuracy will be determined here by the help of the `sklearn.metrics accuracy_score function`. Comparing predicted values `(predictions)` with the corresponding actual labels retrieved from the testing set `y_Test`, this would give an output in the ratio of correct predicted values against all the predicted ones. Accuracy is measured as a good metric in such classification tasks. The result of how well the model is doing is saved in the variable `accuracy`. Finally, the print statement prints the computed accuracy, an `indication of how well the k-NN classifier generalizes to unseen data`.

```python
#Check accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_Test,predictions)
print("Accuracy: ", accuracy)
```

This section of the code is a further assessment of the model's performance as it creates a classification report containing several important metrics other than accuracy, including `precision, recall, and F1-score`, through the use of classification_report from sklearn.metrics, `giving deeper insights into the performance of the model` on the classes.

- `Precision` is the `fraction of positive predictions that are correct`. This measure is useful when the cost of a false positive is high.
- `Recall or sensitivity` is that `portion of real positives well classified by the model`. This arises when false negatives are high.
- `F1-score` is the `harmonic mean of precision and recall` which provides a `balanced measure between them`, particularly when the class distribution is `skewed`.

This function accepts `y_Test as the actual labels` and `predictions as the predicted labels`, printing out these metrics for each class, which, in this case, are the three iris flower species. This can be used to judge not only the general accuracy but also how well the model separates between classes.

```python
#Evaluate Model Performance
from sklearn.metrics import classification_report
print("classification report")
print(classification_report(y_Test,predictions))
```

This code will create a` confusion matrix` for further performance analysis of the k-Nearest Neighbors model. The function from sklearn.metrics used in this line of code, namely confusion_matrix, `compares actual labels y_Test to predicted labels predictions`. A `confusion matrix is a table that aids visualization of how well a classification model is doing when dealing with more than two classes`. The `rows` in this matrix correspond to the `actual class labels`, while the `columns` are the `predicted class labels`. Every entry in the matrix is a `count of how many times a sample belonging to an actual class was predicted` as `belonging to a particular predicted class`.

`Diagonal elements would indicate correct classifications` where the actual class matches the predicted class, while `off-diagonal elements represent misclassifications`. We can see which classes the model is confusing with each other by looking at a confusion matrix, which gives us ideas on where maybe improvement is needed in the model.

```python
#Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_Test,predictions)

print("Confusion Matrix: ")
print(cm)
```

This block of code makes a `scatter plot to visually compare the actual class labels and predicted class labels` on the test data set. It is plotted such that it gives an idea about how well k-NN could classify points in two feature dimensions, namely, sepal length and sepal width. Figure size is set at 8 x 6 inches for better display.

This code executes `a loop through the three classes of iris flowers` and plots all data points in that class by a different color with plt.scatter. This visualizes the results in a scatter plot for actual test data `(X_Test)` using colors derived from the actual class labels, `y_Test`. A second call to scatter `overlays predicted class labels over actual data points`. In this, the `c=predictions` argument plots the points with colors according to the model's predictions, `alpha=0.8` makes the points transparent, and the `cmap="viridis"` makes sure the color scheme holds over the plot. The `xlabel, ylabel, and title functions label the axes and the plot, and plt`.The `legend()` function adds a legend to explain the colors used. Finally, `plt.show()` displays the plot, and based on the first two features of the dataset, there is a visual comparison between the actual and predicted classes.

```python
#Visualization of the Scatter Plot with predicted classes
plt.figure(figsize=(8,6))
for i in range(3):
  plt.scatter(X_Test[:,0][y_Test == i], X_Test[:,1][y_Test == i], cmap='viridis')
plt.scatter(X_Test[:,0],X_Test[:,1], c=predictions, cmap="viridis", alpha=0.8)
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.title("Scatter Plot with Predicted Classes")
plt.legend()
plt.show()
```

Finally, this section of the code will produce a visual representation of the confusion matrix in the form of a `heatmap`. The function `plt.imshow ()` is used in order to produce an image representation of the matrix, with colours representing the values inside the matrix. The color map is set as `'Blues'`. This will assign a blue gradient to the intensity of the value, with deeper shades of blue representing higher values.

The `plt.colorbar()` function draws a` color bar to the side of the heatmap for reference of color intensity in the matrix`. `plt.title("Confusion Matrix")` adds a title to the plot, and `plt.xlabel("Predicted Label")` and `plt.ylabel("True Label")` add labels to the `x-axis and y-axis`, respectively. The` x-axis refers to the class labels predicted`, while the `y-axis refers to the actual class labels`. `plt.xticks()`, and `plt.yticks()` sets tick labels for both the x-axis and the y-axis with `data.target_names` which is an array containing all class names, there are `3 classes` according to `range(3)`, while `data.target_names` give us class names so it would print class names to display its results, then, finally the code shows us `plt.show()`, to print a visual comparison between the true labels against the predicted, enabling one to analyze which each species would get wrong or correct class assignment from the model.

# TEST CODE FOR CODE EXPERIMENTATION

This code imports the required libraries to construct and test a k-Nearest Neighbors (k-NN) model on the Iris dataset, used here as a test code to analyze the study. It uses `matplotlib.pyplot` for visualization, `load_iris` to load the dataset, `train_test_split `to split the data into training and testing sets, `KNeighborsClassifier` to build the k-NN model, and `accuracy_score` to measure the model's accuracy by comparing predicted labels with actual ones. It lets evaluate how well a model performs with the Iris dataset and gives valuable insight into a classification task.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

The following 2 lines of codes were utilized similar to its identical codes from the given code above.

```python
# Load Iris dataset
data = load_iris()
X, y = data.data, data.target
```

This code splits a given dataset into two portions of data with a `80-20 distribution to a train and test dataset respectively`. Using `sklearn.model_selection`, `train_test_split` the dataset splits to form features, X, and the labels y with a specified split being done for a particular random number `random_state=0` that assures repeatability and results from similar executions given different instances and executions.

```python
# Split data into training and testing sets
X_Train, X_Test, y_Train, y_Test = train_test_split(X, y, test_size=0.2, random_state=0)
```

This section of the code attempts various values of `n_neighbors`, or how many neighbors the k-Nearest Neighbors algorithm looks for, and `checks how this affects its selection on the accuracy of the model`. It creates a list called `neighbor_values` with a range of integers where each is of a different number of neighbors to test. An empty list, `accuracies`, is created to hold the accuracy scores for each value of n_neighbors. A `for loop` cycles through each value in neighbor_values, which is a list that contains several steps for each value of k. For each step, a `new k-NN classifier object is created with n_neighbors=k`. The classifier then trains on the training data using the `.fit()` method from the training data `(X_Train, y_Train)`. Finally, the model makes predictions on the test data `(X_Test)` by using the `.predict()` method. The accuracy is calculated by comparing the predicted labels to the `true labels (y_test)` using the `accuracy_score()`. Adding the accuracy score to the accuracies list and `printing result for a single value of n_neighbor shows the k and that k's accuracy`. This helps to determine which is the` best number of neighbors for the k-NN algorithm`, which changes the accuracy as the n_neighbors changes.

```python
# Test different values of n_neighbors
neighbor_values = [1, 5, 10, 15, 33, 50, 66, 70, 80, 90, 100]
accuracies = []

for k in neighbor_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_Train, y_Train)
    predictions = knn.predict(X_Test)
    accuracy = accuracy_score(y_Test, predictions)
    accuracies.append(accuracy)
    print(f"n_neighbors = {k}, Accuracy = {accuracy:.4f}")
```

This part of the code will `plot the outcome of the accuracy` for different n_neighbors values while testing on the k-Nearest Neighbors model. The figure size is taken as 8x5 to see the results more clearly.` plt.plot()` function `plots neighbor_values` against their corresponding accuracies on the y-axis. The plot makes use of `dashed blue lines` with circular markers in `representing the dependency of the accuracy of the model on the number of neighbors (k)`. The xlabel, ylabel, and title functions are utilized in labeling the axes and the plot, therefore providing context for what the graph represents. `xticks()` function is used in ensuring that all the neighbor_values are shown as tick marks on the `x-axis`. Finally, `plt.grid()` adds a grid to the plot to `more easily show the trends as per this function call`. Finally, it calls `plt.show()` in order to produce the plot so that the user can finally see how many neighbors affect the model's accuracy.

```python
# Plot the results
plt.figure(figsize=(8, 5))
plt.plot(neighbor_values, accuracies, marker='o', linestyle='dashed', color='b')
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.title("Effect of n_neighbors on Model Accuracy")
plt.xticks(neighbor_values)
plt.grid()
plt.show()
```
